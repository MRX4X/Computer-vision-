Данный код, позволяет сделать изображение серым

```python
IMG_PATH = '/home/user/ML Sirius/01_opencv/imgs/01_01_Harold.jpg' #Записываем в переменную путь до картинки

img = cv2.imread(IMG_PATH) #С помощью команды imread, считываем изображение

new_img = cv2.resize(img, (300, 300)) #Данная функция, позволяет изменять изображение картинки, с кастомизированными параметрами

new_img = cv2.cvtColor(new_img, cv2.COLOR_BGR2GRAY)

new_img = cv2.Canny(new_img, 30, 30) #Для определения углов, Бинарный формат, все представленно в виде единичек и ноликов

kernel = np.ones((2, 2), np.uint8) # С помощью библиотеки numpy, создаем матрицу размером 5 массивов, в каждом по 5 элементов. uint8 - означает, что числа, большие за ноль только могут использоваться.

new_img = cv2.dilate(new_img, kernel, iterations=1) #Функция dilate, нужна для расширения границ двоичного изображения, пикселей белового плана
                                                    # Функция dilate() принимает следующие параметры.
                                                    # image: это обязательный параметр и исходное изображение, на котором нам нужно выполнить расширение.
                                                    # kernel: Обязательный параметр - матрица, с которой свернуто изображение.
                                                    # iterations: это необязательный параметр, который требует нескольких итераций.

new_img = cv2.erode(new_img, kernel, iterations=1) #Меньшая жирность обводки, нужна для ускорения процесса обработки

cv2.imshow('Harold', new_img) #Данная функция, позволяет открывать картинку в окне. Но также обрезает ее по координатам от нуля (левый верхний угол), от туда идет исчисление, до 100 по ширине и до 150 по длине

cv2.waitKey(0) # Позволяет указать время, сколько будет открыта картинка, время указывается в милисикундах, если стоит значение 0, то данное изображение будет открыто до того времени пока программа не завершит свою работу или пока картинка не закроется
```

Данный код, позволяет отзеркалить изображение

```python
def rotate(img_param, angle): #Два параметра - картинка и угол вращения

    height, wight = img_param.shape[:2] #Срезом берем два значения длины и ширины
    
    point = (wight//2, height//2) #Точка врашения картинки - серина картинки

    mat = cv2.getRotationMatrix2D(point, angle, 1) #Данный метод позволяет вращать изображение. Первый параметр - точка вращения, второй - угол вращения, третий - увеличение изображения по мере вращения, единица - означат, что увеличения не будет.

    return cv2.warpAffine(img_param, mat, (wight, height)) #Возвращает само изображение. Первый парметр - само изображение, второй параметр - матрица вращений, третий параметр - размер изображения

img = rotate(img, 90) #Перезаписываем переменную. Первый параметр - картинка, второй параметр - угол наклона в градусах, если указать отрицательное значение, повернется в другую сторону.

cv2.imshow("Flip", img)

cv2.waitKey(0)
```

Данный код, позволяет повернуть изображение

```python
def transform(img_param, x, y): #Два параметра - смещение по x и по y

    mat = np.float32([[1, 0, x], [0, 1, y]]) #Создаем матрицу, где каждый элемент - число с плавающей точкой

    return cv2.warpAffine(img_param, mat, (img_param.shape[1], img_param.shape[0])) #Возвращает само изображение. Первый парметр - само изображение, второй параметр - матрица вращений, третий параметр - размер изображения по длине и ширине

img = transform(img, 30, 200)

cv2.imshow("Flip", img)

cv2.waitKey(0)
```

Данный код, позволяет сделать смещение фото

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img = cv2.GaussianBlur(img, (5, 5), 0) #Данная функция, позволяет заблюрить изображение, первый параметр - сама картинка, второй параметр - ширина и высота ядра по гаусу, третий параметр - усиленность размытия изображения

img = cv2.Canny(img, 35, 35) #Для определения углов, Бинарный формат, все представленно в виде единичек и ноликов

con, hir = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE) #Контуры можно объяснить просто как кривую, соединяющую все непрерывные точки (вдоль границы), имеющие одинаковый цвет или интенсивность. Контуры — полезный инструмент для анализа формы, обнаружения и распознавания объектов. RETR_LIST - позволяет получать все возможные контуры. CHAIN_APPROX_NONE - позволяет получать все возможные точеки контура

print(con, hir) #Первая переменная con - содержит все списки контуров объектов, hir - порядок (иерархия) контуров объектов

cv2.imshow("Flip", img)

cv2.waitKey(0)
```

Данный код, позволяет выделить контуры изображения

```python
img = cv2.imread("/home/user/ML Sirius/01_opencv/imgs/02_01_Keanu.jpg")

new_img = np.zeros(img.shape, dtype="uint8") #Создание черного изображения, т.к ноль - черный цвет

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img = cv2.GaussianBlur(img, (5, 5), 0)

img = cv2.Canny(img, 35, 35)

con, hir = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

cv2.drawContours(new_img, con, -1, (230, 111, 148), 1) #Метод создания изображения на основе контуров

print(con, hir)

cv2.imshow("Flip", new_img)

cv2.waitKey(0)
```
